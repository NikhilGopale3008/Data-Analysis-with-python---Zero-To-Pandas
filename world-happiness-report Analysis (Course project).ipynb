## Analyze Data

## 3) Lets find Largest 20 country 'Healthy life expectancy at birth'

## Rename Column name

```python
world_happy = world_happy.rename(columns={'Country name':'Country_name'})
```

## Find The Largest Healty countries 

```python
largest_healty_countries = world_happy.nlargest(20, 'Healthy life expectancy at birth')
largest_healty_countries
```

```python
largest_healty_countriess = largest_healty_countries[largest_healty_countries.Country_name == 'Singapore']
largest_healty_countriess
```

## As you can see Singapore is the most healthy country

```python
larg_singapore = largest_healty_countries['Healthy life expectancy at birth'].mean()
print('The average of singapores  Healthy life expectancy at birth is {}'.format(larg_singapore))
```

## 4)Lets find lowest 20 country 'Healthy life expectancy at birth

## Find The Lowest Healty countries

```python
lowest_healty_countries = world_happy.nsmallest(20, 'Healthy life expectancy at birth')
lowest_healty_countries
```

## As you can see 	Hong Kong S.A.R. of China  is the most healthy country

```python
lowest_healty_countriess = lowest_healty_countries[lowest_healty_countries.Country_name == 'Hong Kong S.A.R. of China']
lowest_healty_countriess
```

```python
low_health= lowest_healty_countries['Healthy life expectancy at birth'].mean()
print('The average of Hong Kong S.A.R. of China  Healthy life expectancy at birth is {}'.format(low_health))
```
